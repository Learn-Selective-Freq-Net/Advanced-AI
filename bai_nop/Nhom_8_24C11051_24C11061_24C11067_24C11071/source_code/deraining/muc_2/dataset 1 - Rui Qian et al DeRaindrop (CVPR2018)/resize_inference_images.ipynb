```python
gt_path_a = "./ground_truth/raindrop_data/test_a/"
inference_data_path_a = "./inference_result/SFNet/deraining/test_a/"

gt_path_b = "./ground_truth/raindrop_data/test_b/"
inference_data_path_b = "./inference_result/SFNet/deraining/test_b/"
```

```python
resized_inference_data_dir = "./resized_inference_data/"
```

```python
all_dirs = [
    (gt_path_a, inference_data_path_a),
    (gt_path_b, inference_data_path_b)
]
```

```python
import os
from PIL import Image

# Resize inference image to ground truth image's size
for gt_path, inference_data_path in all_dirs:
    for file in os.listdir(gt_path):
        if file.endswith(".png") or file.endswith(".jpg"):
            gt_file = os.path.join(gt_path, file)
            inference_file = os.path.join(inference_data_path, file)
            print("Processing", gt_file, inference_file)

            gt_image = Image.open(gt_file)
            inference_image = Image.open(inference_file)
            inference_image = inference_image.resize(gt_image.size)
            
            resized_image_filepath = os.path.join(
                resized_inference_data_dir, 
                gt_path.split("/")[-2], 
                file
            )

            os.makedirs(os.path.dirname(resized_image_filepath), exist_ok=True)
            inference_image.save(resized_image_filepath)
            print("Saved to", resized_image_filepath)
```
