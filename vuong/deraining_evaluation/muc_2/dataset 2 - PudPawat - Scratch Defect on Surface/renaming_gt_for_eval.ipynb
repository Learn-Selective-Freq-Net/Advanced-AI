```python
import os

defect_dir = "./defect/"
selected_non_defect_dir = "./non defect selection/"
selected_non_defect_renamed_dir = "./non defect selection renamed/"
```

```python
def get_files(dir):
    files = []
    for root, dirs, filenames in os.walk(dir):
        for filename in filenames:
            files.append(os.path.join(root, filename))
    return files
```

```python
defect_files = get_files(defect_dir)
selected_non_defect_files = get_files(selected_non_defect_dir)
```

```python
import shutil

if len(defect_files) != len(selected_non_defect_files):
    print("Number of defect and non-defect files do not match:", len(defect_files), len(selected_non_defect_files))
else:
    for i in range(len(defect_files)):
        defect_file = defect_files[i]
        selected_non_defect_file = selected_non_defect_files[i]

        defect_filename = os.path.basename(defect_file)
        renamed_filename = selected_non_defect_renamed_dir + defect_filename

        # Copy the non-defect file to the renamed directory
        shutil.copyfile(selected_non_defect_file, renamed_filename)

        print("Renamed", selected_non_defect_file, "to", renamed_filename)
```
