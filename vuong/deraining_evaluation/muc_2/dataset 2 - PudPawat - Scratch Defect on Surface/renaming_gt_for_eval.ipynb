```python
import os

defect_dir = "./defect"
```

```python
def get_files(dir):
    files = []
    for root, dirs, filenames in os.walk(dir):
        for filename in filenames:
            files.append(os.path.join(root, filename))
    return files
```

```python
defect_files = get_files(defect_dir)
```
