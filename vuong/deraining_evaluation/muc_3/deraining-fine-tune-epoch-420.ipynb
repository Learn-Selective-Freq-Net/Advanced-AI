```python
import os

# Ensuring corresponding target image name for each input image
folder_paths = [
    # "../raindrop_data/test_a/",
    # "../raindrop_data/test_b/",
    "/kaggle/working/train_data/input/",
    "/kaggle/working/train_data/target/",
    "/kaggle/working/valid_data/input/",
    "/kaggle/working/valid_data/target/",
]

i = 0
n = 10

for folder_path in folder_paths:
    ext = "png" if (
        ("train" in folder_path) 
        or ("valid" in folder_path)
    ) else "jpg"
    
    for filename in os.listdir(folder_path):
        postfix = "_clean" if "target" in folder_path else "_rain"
        
        if filename.endswith(f"{postfix}.{ext}"):
            new_name = filename.replace(f"{postfix}.{ext}", f".{ext}")
            old_path = os.path.join(folder_path, filename)
            new_path = os.path.join(folder_path, new_name)
            os.rename(old_path, new_path)

            # Too much data to display
            i = i + 1
            if i % n == 0:
                print(f'Renamed: {old_path} -> {new_path}')
```
